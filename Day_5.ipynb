Function to Check Even or Odd

```python
# Function to check even or odd

def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

number = int(input("Enter a number: "))

print(check_even_odd(number))
```

Function to Find the Largest of Three Numbers

```python
# Function to find the largest number

def largest(a, b, c):
    return max(a, b, c)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

print("Largest =", largest(a, b, c))
```

Function to Calculate Factoria

```python
# Function to calculate factorial

def factorial(num):
    fact = 1

    for i in range(1, num + 1):
        fact *= i

    return fact

number = int(input("Enter a number: "))

print("Factorial =", factorial(number))
```

Function to Check Prime Number

```python
# Function to check prime number

def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True

number = int(input("Enter a number: "))

if is_prime(number):
    print("Prime Number")
else:
    print("Not Prime Number")
```

Function to Reverse a String

```python
# Function to reverse a string

def reverse_string(text):
    return text[::-1]

string = input("Enter a string: ")

print("Reversed String:", reverse_string(string))
```

Function to Count Vowels

```python
# Function to count vowels

def count_vowels(text):
    count = 0

    for ch in text.lower():
        if ch in "aeiou":
            count += 1

    return count

string = input("Enter a string: ")

print("Number of vowels =", count_vowels(string))
```

Function to Calculate the Average of Numbers

```python
# Function to calculate average

def average(numbers):
    return sum(numbers) / len(numbers)

numbers = []

n = int(input("How many numbers? "))

for i in range(n):
    num = float(input("Enter number: "))
    numbers.append(num)

print("Average =", average(numbers))
```

Recursive Function to Find Factorial

```python
# Recursive function to find factorial

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter a number: "))

print("Factorial =", factorial(num))
```

Recursive Function to Print Fibonacci Series

```python
# Recursive Fibonacci function

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

terms = int(input("Enter number of terms: "))

for i in range(terms):
    print(fibonacci(i), end=" ")
```

Function to Check Palindrome

```python
# Function to check palindrome

def is_palindrome(text):
    return text == text[::-1]

word = input("Enter a word: ")

if is_palindrome(word):
    print("Palindrome")
else:
    print("Not Palindrome")
```

Function to Count Uppercase and Lowercase Letters

```python
# Count uppercase and lowercase letters

def count_letters(text):
    upper = 0
    lower = 0

    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Uppercase Letters:", upper)
    print("Lowercase Letters:", lower)

string = input("Enter a string: ")

count_letters(string)
```

Function to Remove Duplicates from a List

```python
# Remove duplicate elements

def remove_duplicates(items):
    return list(set(items))

numbers = [1, 2, 3, 2, 4, 5, 4, 6]

print("Original List:", numbers)
print("Without Duplicates:", remove_duplicates(numbers))
```

Function to Check Armstrong Number

```python
# Armstrong number

def armstrong(num):
    digits = len(str(num))
    total = sum(int(digit) ** digits for digit in str(num))
    return total == num

number = int(input("Enter a number: "))

if armstrong(number):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

Function to Calculate Power

```python
# Calculate power

def power(base, exponent):
    return base ** exponent

base = int(input("Enter base: "))
exp = int(input("Enter exponent: "))

print("Result =", power(base, exp))
```
