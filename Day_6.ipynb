Check Prime Number

```python
# Program 1: Check Prime Number

num = int(input("Enter a number: "))

if num < 2:
    print(num, "is not a prime number.")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, "is a prime number.")
    else:
        print(num, "is not a prime number.")
```

Generate Prime Numbers in a Range

```python
# Program 2: Prime Numbers in a Range

start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))

print("Prime numbers are:")

for num in range(start, end + 1):
    if num > 1:
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
```

Function to Find Maximum Number

```python
# Program 3: Function to Find Maximum Number

def maximum(a, b, c):
    return max(a, b, c)

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

print("Largest number is:", maximum(num1, num2, num3))
```

Count Vowels in a String

```python
# Program 4: Count Vowels

text = input("Enter a string: ")

count = 0
vowels = "aeiouAEIOU"

for char in text:
    if char in vowels:
        count += 1

print("Total vowels:", count)
```

Reverse a String Using Function

```python
# Program 5: Reverse String

def reverse_string(text):
    return text[::-1]

string = input("Enter a string: ")

print("Reversed string:", reverse_string(string))
```

Recursive Sum of Natural Numbers

```python
# Program 6: Sum Using Recursion

def natural_sum(n):
    if n == 0:
        return 0
    return n + natural_sum(n - 1)

number = int(input("Enter a positive number: "))

print("Sum =", natural_sum(number))
```

Lambda Function for Cube

```python
# Program 7: Lambda Function

cube = lambda x: x ** 3

number = int(input("Enter a number: "))

print("Cube =", cube(number))
```

Map Function to Convert Celsius to Fahrenheit

```python
# Program 8: Map Function

celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

Filter Positive Numbers

```python
# Program 9: Filter Positive Numbers

numbers = [-10, 15, -20, 30, -40, 50]

positive = list(filter(lambda x: x > 0, numbers))

print("Positive numbers:", positive)
```

Reduce Function to Find Product of List

```python
# Program 10: Reduce Function

from functools import reduce

numbers = [1, 2, 3, 4, 5]

product = reduce(lambda x, y: x * y, numbers)

print("Product of list:", product)
```

Exception Handling (Division)

```python
# Program 15: Exception Handling

try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    result = num1 / num2

    print("Result:", result)

except ZeroDivisionError:
    print("Cannot divide by zero.")

except ValueError:
    print("Please enter valid integers.")
```
