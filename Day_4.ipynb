Check Prime Number

```python


number = int(input("Enter a number: "))

is_prime = True

if number <= 1:
    is_prime = False
else:
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
            break

if is_prime:
    print(number, "is a Prime Number")
else:
    print(number, "is NOT a Prime Number")
```

Fibonacci Series

```python
# ============================================
# Day 4 - Program 19
# Fibonacci series
# ============================================

terms = int(input("Enter number of terms: "))

first = 0
second = 1

print("Fibonacci Series:")

for i in range(terms):
    print(first, end=" ")
    next_number = first + second
    first = second
    second = next_number
```

Count Positive, Negative, and Zero Numbers

```python
# ============================================
# Day 4 - Program 22
# Count positive, negative and zero numbers
# ============================================

positive = 0
negative = 0
zero = 0

for i in range(5):
    number = int(input(f"Enter number {i + 1}: "))

    if number > 0:
        positive += 1
    elif number < 0:
        negative += 1
    else:
        zero += 1

print("\nResults")
print("Positive Numbers:", positive)
print("Negative Numbers:", negative)
print("Zero Numbers:", zero)
```

Find the Largest Number

```python
# ============================================
# Day 4 - Program 23
# Find the largest number
# ============================================

largest = None

for i in range(5):
    number = int(input(f"Enter number {i + 1}: "))

    if largest is None or number > largest:
        largest = number

print("\nLargest Number =", largest)
```

Print a Number Pyramid

```python


rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

Login System (Maximum 3 Attempts)

```python
# ============================================
# Day 4 - Program 25
# Login with maximum 3 attempts
# ============================================

correct_password = "python123"

for attempt in range(1, 4):
    password = input("Enter Password: ")

    if password == correct_password:
        print("Login Successful!")
        break
    else:
        print("Incorrect Password!")

else:
    print("Your account has been locked after 3 failed attempts.")
```
