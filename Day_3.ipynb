Creating Strings

```python
# Creating Strings

name = "Naveed Khan"
city = 'Peshawar'
course = """Python Programming"""

print(name)
print(city)
print(course)
```

Data Type of String

```python
text = "Artificial Intelligence"

print(text)
print(type(text))
```

String Indexing

```python
language = "Python"

print("First Character :", language[0])
print("Second Character :", language[1])
print("Third Character :", language[2])
print("Fourth Character :", language[3])
print("Fifth Character :", language[4])
print("Sixth Character :", language[5])
```

Negative Indexing

```python
language = "Python"

print("Last Character :", language[-1])
print("Second Last :", language[-2])
print("Third Last :", language[-3])
print("Fourth Last :", language[-4])
print("Fifth Last :", language[-5])
print("Sixth Last :", language[-6])
```

String Slicing

```python
text = "Artificial Intelligence"

print(text[0:10])
print(text[11:23])
print(text[:10])
print(text[11:])
print(text[:])
```

Length of String

```python
message = "Welcome to Python"

print(message)

print("Length =", len(message))
```

String Concatenation

```python
first_name = "Naveed"
last_name = "Khan"

full_name = first_name + " " + last_name

print("Full Name:", full_name)
```


String Repetition

```python
word = "Python "

print(word * 5)
```

Escape Characters

```python
print("Hello\nWorld")

print("Python\tProgramming")

print("He said, \"Python is easy.\"")

print("This is a backslash: \\")
```

Membership Operators (in & not in)

```python
text = "Python Programming"

print("Python" in text)
print("Java" in text)

print("Programming" in text)
print("AI" not in text)
```

Loop Through a String

```python
language = "Python"

print("Characters in the string:")

for character in language:
    print(character)
```

Reverse a String

```python
text = "Python"

reverse = text[::-1]

print("Original String :", text)
print("Reversed String :", reverse)
```

First and Last Character

```python
text = input("Enter a word: ")

print("First Character :", text[0])
print("Last Character :", text[-1])
```

Count Number of Characters

```python
sentence = input("Enter a sentence: ")

print("Total Characters =", len(sentence))
```

Check Palindrome

```python
word = input("Enter a word: ")

if word == word[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

Check a Word Exists

```python
sentence = input("Enter a sentence: ")

search = input("Enter a word to search: ")

if search in sentence:
    print("Word Found")
else:
    print("Word Not Found")
```

Display Initials

```python
first = input("Enter First Name: ")
last = input("Enter Last Name: ")

print("Initials =", first[0].upper() + "." + last[0].upper())
```

Greeting Program (f-String)

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print(f"Hello {name}! You are {age} years old.")
print(f"Welcome to Python Programming, {name}.")
```

Mini Username Generator

```python
first = input("First Name: ")
last = input("Last Name: ")

username = first.lower() + "_" + last.lower()

print("Generated Username:", username)
```

Convert String to Uppercase

```python
text = input("Enter a sentence: ")

print("Original :", text)
print("Uppercase :", text.upper())
```

Lowercase, Title Case & Capitalize

```python
text = input("Enter your full name: ")

print("Lowercase :", text.lower())
print("Title Case :", text.title())
print("Capitalize :", text.capitalize())
```

Replace a Word

```python
sentence = input("Enter a sentence: ")

old_word = input("Word to replace: ")
new_word = input("New word: ")

updated = sentence.replace(old_word, new_word)

print("Updated Sentence:")
print(updated)
```

Split a String

```python
sentence = input("Enter a sentence: ")

words = sentence.split()

print("Words in the sentence:")

for word in words:
    print(word)
```

Remove Extra Spaces

```python
text = input("Enter a sentence with extra spaces: ")

print("Before:")
print(text)

print("After:")
print(text.strip())
```

Find a Word

```python
sentence = input("Enter a sentence: ")

word = input("Enter word to find: ")

position = sentence.find(word)

if position != -1:
    print("Word found at index:", position)
else:
    print("Word not found.")
```

Count Occurrences of a Character

```python
text = input("Enter a sentence: ")

character = input("Enter a character: ")

print("Occurrences:", text.count(character))
```

Mini String Analyzer Project

```python
sentence = input("Enter a sentence: ")

print("\n========== STRING ANALYZER ==========")

print("Original Sentence :", sentence)
print("Length :", len(sentence))
print("Uppercase :", sentence.upper())
print("Lowercase :", sentence.lower())
print("Title Case :", sentence.title())
print("First Character :", sentence[0])
print("Last Character :", sentence[-1])
print("Number of Words :", len(sentence.split()))
print("Reversed :", sentence[::-1])
```
